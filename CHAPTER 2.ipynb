# **2.1 전도 전달률 방정식 ( Fourier's Law )**

### -열전달률(Heat rate)

**<p style = "font-size: 20px; line-height : 1.2em"> $$ q_x = -kA{dT \over dx } (단위 : [W]) $$</p>**

### -열유속(Heat flux)

**<p style = "font-size: 20px; line-height : 1.2em"> $$ q^{''}_x = {q_x \over A} = -k{dT \over dx} (단위 : [W/m^2])$$</p>**

> 열흐름의 방향은 항상 **등온(isothermal)** 표면에 수직인 방향이다.




### -일반적인 전도 전달률 방정식

**<p style = "font-size: 20px; line-height : 1.2em"> $$ \overrightarrow{q^{''}} = -k \nabla T $$</p>**
> 열유속 벡터는 **등온(isothermal)** 표면에 수직이다.  
> 전도가 일어나는 매질을 **등방성(isotropic)** 으로 가정한다.


## 2.1.1 <font color = red>**직교좌표계**</font>에서의 일반적인 전도 전달률 방정식


**<p style = "font-size: 20px; line-height : 1.2em"> $$q^{''} = - k \nabla T(x,y,z) = -k{ \frac{\partial T}{\partial x}\vec{i} - k\frac{\partial T}{\partial y}\vec{j} - k \frac{\partial T}{\partial z}}\vec{k}$$</p>**


## 2.1.2 <font color = red>**원통좌표계**</font> 에서의 열유속 벡터의 일반적인 형태와 Fourier법칙

**<p style = "font-size: 20px; line-height : 1.2em"> $$q^{''} = - k \nabla T(r, \phi , z) =  -k{ \frac{\partial T}{\partial r}\vec{i} - k\frac{\partial T}{r\partial \phi}\vec{j} - k \frac{\partial T}{\partial z}}\vec{k}$$ </p>**



---

## 2.1.3 <font color = red>**구좌표계**</font> 에서의 열유속 벡터의 일반적인 형태와 Fourier법칙 

**<p style = "font-size: 20px; line-height : 1.2em"> $$q^{''} = - k \nabla T(r, \phi , \theta) =  -k{ \frac{\partial T}{\partial r}\vec{i} - k\frac{\partial T}{r\partial \phi}\vec{j} - k \frac{\partial T}{r\partial \theta}}\vec{k}$$ </p>**

# **2.3 열확산 방정식**

##  2.3.1  <font color = red>**직교좌표계**</font>로 표시된 열확산 방정식의 일반형태인 열방정식 ( Heat equation )

**<p style = "font-size: 18px; line-height : 1.2em"> $$ \frac{\partial}{\partial x}(k\frac{\partial T}{\partial x}) + \frac{\partial}{\partial y}(k\frac{\partial T}{\partial y}) + \frac{\partial}{\partial z}(k\frac{\partial T}{\partial z}) + \dot{q} = \rho c_{p} \frac{\partial T}{\partial t}$$ </p>**



## 2.3.2 <font color = red>**원통좌표계**</font>로 표시된 열확산 방정식의 일반형태인 열방정식 ( Heat equation )

**<p style = "font-size: 18px; line-height : 1.2em"> $$ \frac{1}{r}\frac{\partial}{\partial r}(kr\frac{\partial T}{\partial r}) + \frac{1}{r^{2}}\frac{\partial}{\partial \phi}(k\frac{\partial T}{\partial \phi}) + \frac{\partial}{\partial z}(k\frac{\partial T}{\partial z}) + \dot{q} = \rho c_{p} \frac{\partial T}{\partial t}$$ </p>**



## 2.3.3 <font color = red>**구좌표**</font>로 표시된 열확산 방정식의 일반형태인 열방정식 ( Heat equation )

**<p style = "font-size: 18px; line-height : 1.2em"> $$ \frac{1}{r^{2}}\frac{\partial}{\partial r}(kr^{2}\frac{\partial T}{\partial r}) + \frac{1}{r^{2}sin^{2}\theta}\frac{\partial}{\partial \phi}(k\frac{\partial T}{\partial \phi}) + \frac{1}{r^{2}sin^{2}\theta}\frac{\partial}{\partial \theta}(k\frac{\partial T}{\partial \theta}) + \dot{q} = \rho c_{p} \frac{\partial T}{\partial t}$$ </p>**



# **2.4 경계조건 과 초기조건**
