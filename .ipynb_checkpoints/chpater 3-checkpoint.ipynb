1차원 이라는 말은 종속변수들의 공간적 변화를 나타내기위해 단지 하나의 좌표만 필요하다는 것이다.
따라서 1차원 계에서 온도구배는 단지 하나의 좌표방향에 따라서만 존재하고 열전달은 그 방향으로만 일어난다.


## 3.1 평면벽

### 3.1.1 온도분포



### 열방정식(Heat equation)
> 벽 내부에 분포된 열원이나 열침이 없는 정상상태 조건


### $${d \over dx}(k {dT \over dx})=0$$ 


> 열발생이 없는 평면벽에서 1차원 정상상태 전도에 대하여 열유속은 일정하며 x와 무관하다.   
> 위의 방정식을 두번 적분하여 다음과 같은 일반해를 얻을 수 있다.

### 일반해(General solution)

### $$ T(x) = C_1x + C_2 $$

>**경계조건(Boundary Condition)** : $ T(0) = T_{s,1}, T(L) = T_{s,2} $ 



### 평면벽에서 온도분포 


### $$ T(x) = (T_{s,2} - T_{s,1}){x \over L} + T_{s,1} $$



>열발생이 없고, 열전도율이 일정한 평면벽 내에서의 1차원 정상상태 전도일 때 온도는 x에 따라 선형적으로 변한다는 것을 알 수 있다.


### 전도 열전달률

 
### $$ q_x = -kA{dT \over dx} = {kA \over L}(T_{s,1}- T_{s,2}) $$
 
 
 
>여기서 A는 열전달 방향에 수직인 벽의 면적이며, 평면벽에서는 x에 무관하게 일정하다는 것에 유의해야 한다. 

### 열유속

 
### $$ q''_x = {q_x \over A} = { k \over L}(T_{s,1} - T_{s,2}) $$
 
 

> 위 두식은 열전달률 qx와 열유속 q''x이 x에 관계없이 일정하다는 것을 나타낸다.

## 3.1.2 열저항

### 전도 열저항

 
### $$ R_{t,cond} = {T_{s,1} - T_{s,2} \over q_x} = { L \over kA } (단위 : [K/W])$$
 
> ### 단위 면적당 전도 열저항 : $ R^{''}_{t,cond} = {L \over k} (단위 : [m^2K/W])$  

### 대류 열저항


### $$ R_{f,conv} = { T_s - T_\infty \over q } = { 1 \over hA } $$

> ### 단위 면적당 대류 열저항 : $ R^{''}_{t,conv} = {1 \over h} (단위 : [m^2K/W])$ 


### 평면벽의 등가 열회로




### $$ R_{tot} = { 1 \over h_1A } + { L \over kA } + { 1 \over h_2A } $$

### $$ q_x = {T_{\infty,1}-T_{\infty,2}\over R_{tot}} $$ 
 
 
### 복사 열저항


### $$ R_{t,rad} = { T_s - T_{sur} \over q_{rad} } = { 1 \over h_rA } $$

> ### 단위 면적당 복사 열저항 : $ R^{''}_{t,rad} = {1 \over h_r} (단위 : [m^2K/W])$    
> ### $ h_r = \epsilon\sigma(T_s + T_{sur})(T^2_s + T^2_{sur}) $

## 3.1.3 복합벽

> 복합벽과 같은 더 복잡한 계에서도 등가 열회로를 사용할 수 있다. 




### 이 계에 대한 1차원 열전달률.

> 접촉 열저항은 무시한다.
### $$ q_x = { T_{\infty,1} - T_{\infty,4}  \over \Sigma R_t} $$

### $$ \Sigma R_t = R_{tot} = {1 \over A}[{1 \over h_1}+{L_A \over k_A}+{L_B \over k_B}+{L_c \over k_C}+{1 \over h_4}]$$


### 열관류계수(Overall Heat Transfer Coefficient) U   


### $$ q_x = UA \Delta T_{overall} $$

### $$ U = { 1 \over R_{tot}A } = { 1\over [( 1/h_1 ) + (L_A/k_A ) + (L_B/k_B) + (L_C/ k_c ) + (1/ h_4) ] }$$
 

### $$ R_{tot} = {\sum}R_t = {{\Delta}T \over q} = {1 \over UA} $$

### **그림 3.3 직렬-병렬 복합벽에 대한 등가 열회로**



> #### Note departure from one-dimensional conditions for $ k_F \neq k_G$

**(b)에서는 x 방향에 평행한 표면이 단열이라고 가정한다.**



## 3.1.4 접촉저항

### $$ R_{t,c} = {R{''}_{t,c} \over A_c} $$

> ### Values depend on  
   > * A 와 B 물질
   > * 표면 거칠기
   > * 두 물질 사이 틈새 조건
   > * 접촉 압력

## 3.1.5 다공성 물질


>많은 응용에서 열전달이 비유동성 고체와 유체의 결합으로 이루어진 다공성 물질 내에서 발생하게 된다.  
x=0에서의 표면온도 T1, x=L 에서의 표면온도 T2의 포화된 다공성 물질에 대해 고려해보자.  

### 열전달률
>정상상태에 도달한 이후에, 만약 T1 > T2일 경우
### $$ q_x = {k_{eff}A \over L}(T_1-T_2) $$
 
 
Maxwell의 결과는 구형 유체영역의 희석 분포가 존재하는 고체상의 연결구조로 구성된 포화 다공성 물질의 유효 열전도율을 구하는데 이용할 수 있다. 

### $$ k_{eff} = [{k_f + 2K_s - 2{\epsilon}(k_s - k_f) \over k_f + 2k_s +  {\epsilon}(k_s - k_f)}]k_s $$
>$K_{eff}$ : 유효열전도율   
>상대적으로 작은 공극률$ (ε \leq 0.25) $에 대하여 유효하다. 


```python
"""
example 3-1
"""

import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
from scipy.constants import sigma
from scipy.optimize import fsolve

K0 = 273.15
qmax = 100 # maximum heat loss (metabolic heat generation rate), W
Lsf = 3e-3 # fat thickness
ksf, kins = 0.3, 0.014
A = 1.8 # human surface area
Ti, Tinf, Tsur = 35+K0, 10+K0, 10+K0

# air
h, epsilon = 2, 0.95
def fun(x, h, epsilon):
    T, Lins = x
    qk = (Ti - T)/(Lsf/(ksf*A) + Lins/(kins*A))
    qh = (T - Tinf)/(1/(h*A))
    qr = sigma*epsilon*A*(T**4 - Tsur**4)
    return [qmax - (qh + qr), qk - qmax]

x0 = [300, 5e-3]
T, Lins = fsolve(fun, x0, args=(h, epsilon))
Ts = Ti - qmax*(Lsf/(ksf*A))
qh = h*A*(T - Tinf)
qr = sigma*epsilon*A*(T**4 - Tsur**4)
hr = qr/(A*(T - Tsur))
print('\nair')
print('insulation, mm   : {:10.2f}'.format(Lins*1000))
print('surface temp., C : {:10.2f}'.format(T - K0))
print('skin temp., C    : {:10.2f}'.format(Ts - K0))
print('convection, W    : {:10.2f}'.format(qh))
print('radiation, W     : {:10.2f}'.format(qr))
print('hr, W/m2-K       : {:10.2f}'.format(hr))

# water
h, epsilon = 200, 0
T, Lins = fsolve(fun, x0, args=(h, epsilon))
print('\nwater')
print('insulation, mm   : {:10.2f}'.format(Lins*1000))
print('surface temp., C : {:10.2f}'.format(T - K0))

# 
h = np.geomspace(2, 100)
for epsilon in [1, 0.8, 0.6, 0.4, 0.2]:
    f = np.array([fsolve(fun, x0, args=(h1, epsilon)) for h1 in h])
    plt.semilogx(h, f[:,1]*1000, label=epsilon)
    
plt.xlabel(r'h, W/m$^2$-K')    
plt.ylabel(r'L$_{ins}$, mm')    
d = np.arange(1,10)
plt.xticks(np.r_[d, d*10, 100])
plt.xlim([2,100])
plt.legend(title=r'$\epsilon$')
plt.show()
```

```python
"""
example 3-2
"""
import numpy as np

Tinf = 25
L, k = 8e-3, 239 # thickness and thermal conductivity of aluminum substrate
Rtc = 0.9e-4 # thermal contact resistance, m2-K/W
#%% a
qc = 1e4 # heat dissipation from chip, W/m2
h = 100
Rtop = 1/h
Rbot = Rtc + L/k + 1/h
R = 1/(1/Rtop + 1/Rbot)
print(Rtop, Rbot, R)
Tc = Tinf + qc*R
print(Tc)
#%% top blockage
Tc = Tinf + qc*Rbot
print(Tc)
#%% bottom blockage
Tc = Tinf + qc*Rtop
print(Tc)

#%%
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

# allowable chip temp. 85 C
Tc = 85
h = np.linspace(100, 2000)
Rtop = 1/h
Rbot = Rtc + L/k + 1/h
qc = (Tc - Tinf)*(1/Rtop + 1/Rbot)

plt.plot(h, qc/1e4)
plt.xlabel(r'h, W/m$^2$-K')    
plt.ylabel(r'$q_c$, W/cm$^2$')
plt.show()
```

```python
"""
example 3-3
"""
import numpy as np
from scipy.constants import sigma
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
from scipy.optimize import fsolve

K0 = 273.15
L, W = 1, 0.1
A = L*W 
Lg, La = 3e-3, 0.1e-3 # glass, adhesive
kg, ka = 1.4, 145
Tinf, Tsur = 20, 20
h = 35 # convection coefficient
Gs = 700 # irradiation, W/m2
# emissivity
epsilon = 0.9
# reflectivity, absorptivity
rho, alpha = 0.07, 0.1
tau = 1 - rho - alpha # transmisivity

def eta_fun(Tsi):
    """ solar - electrical conversion efficiency """
    return 0.553 - 0.001*(Tsi + K0)

def fun(x,h):
    Tsi, Ttop = x    
    eta = eta_fun(Tsi)
    g1 = tau*Gs*A*(1 - eta)
    q1 = (Tsi - Ttop)/(Lg/(kg*A) + La/(ka*A))
    g2 = alpha*Gs*A 
    q3h = h*A*(Ttop - Tinf)
    q3r = sigma*epsilon*A*((Ttop+K0)**4 - (Tsur+K0)**4)
    return [g1 - q1, (g1 + g2) - (q3h + q3r)]

Tsi, Ttop = fsolve(fun, [33, 32], args=(h,))
eta = eta_fun(Tsi)
P = tau*Gs*A*eta 
print('power, W : ', P)
print('efficiency : ', eta)
print('T_si, T_top : ', Tsi, Ttop)

#%% effect of h
h = np.linspace(15, 55)
f = np.array([fsolve(fun, [33, 32], args=(h1,)) for h1 in h])
Tsi, Ttop = f[:,0], f[:,1]
eta = eta_fun(Tsi)
P = tau*Gs*A*eta 
plt.plot(h, P)
plt.xlabel('h, W/m2-K')
plt.ylabel('P, W')    
plt.show()    

#%% concentrating photovoltaic system
h = 55
Gs = 7000
Tsi, Ttop = fsolve(fun, [33, 32], args=(h,))
eta = eta_fun(Tsi)
P = tau*Gs*A*eta 
print(P, eta, Tsi, Ttop)

q3h = h*A*(Ttop - Tinf)
q3r = sigma*epsilon*A*((Ttop+K0)**4 - (Tsur+K0)**4)
print(q3h, q3r)
```

```python
"""
example 3-4
thermal conductivity of carbon nanotube
"""

import numpy as np

D = 14e-9 # dia. of carbon nanotube
s = 5e-6  # length of carbon nanotube 
k_pt = 71.6 # platinum
k_sn = 15.5 # silicon nitride
L_sn = 250e-6
L_pt = L_sn
q = 11.3e-6
T_s = 308.4 # temp. of sensing island
T_inf = 300 # temp. of silicon nitride block
w_sn, t_sn = 3e-6, 0.5e-6
w_pt, t_pt = 1e-6, 0.2e-6

A_pt = w_pt*t_pt
A_sn = w_sn*t_sn - A_pt
A_cn = np.pi*D**2/4

R_sn = L_sn/(k_sn*A_sn)
R_pt = L_pt/(k_pt*A_pt)
R_sup = 1/(1/R_sn + 1/R_pt)

q_s = 2*(T_s - T_inf)/R_sup
q_h = q - q_s
T_h = T_inf + (q_h/2)*R_sup # temp. of heated island

k_cn = q_s*s/(A_cn*(T_h - T_s))
print(T_h, T_s, k_cn)

# radiation 
T_sur = 300
sigma = 5.67e-8
q_rad = sigma*2*(10e-6)**2*(T_h**4 - T_sur**4)
```

## 3.2 전도해석의 대체 풀이방법

## 3.3 반지름 방향 계
>원통형 계와 구형 계에서는 종종 반지름 방향으로만의 온도구배가 형성되므로, 이때는 1차원으로 취급할 수 있다. 

## 3.3.1 원통

>일반적인 예로서 내부표면과 외부표면이 온도가 서로 다른 유체에 노출되어있는 속이 빈 원통이 있다. 



### 열방정식(Heat equation)
 >열발생이 없는 정상상태 조건
 
### $$ {1 \over r}{d \over dr}(kr{dT \over dr}) = 0$$

### 일반해(General solution)
> k 값이 일정하다고 가정한다
### $$ T(r) = C_1lnr + C_2 $$

>**경계조건(Boundary Condition)** : $ T(r_1) = T_{s,1}, T(r_2) = T_{s,2} $ 

### 원통에서 온도분포
.

### $$ T(r) = {T_{s,1} - T_{s,2} \over ln(r_1/r_2)}ln({r \over r_2}) + T_{s,2}$$


### 열전달률


### $$ q_r = -kA{dT \over dr} = -k(2{\pi}rL){dT \over dr} $$ 
> 열전달률은 반지름 방향으로 일정하다.

### $$ q_r =  2 \pi rLq^{''}_r = {2{\pi}Lk(T_{s,1} - T_{s,2}) \over ln(r_2/r_1)} (단위:[W])$$

### 열유속
### $$ q^{''}_r = -k{dT \over dr} = {k(T_{s,1} - T_{s,2}) \over rln(r_2/r_1)} (단위:[W/m^2])$$
### $$ q^{'}_r = 2 \pi rq^{''}_r = {2{\pi}k(T_{s,1} - T_{s,2}) \over ln(r_2/r_1)} (단위:[W/m])$$

### 반지름방향 전도에 대한 열저항

### $$ R_{t,cond} = {ln(r_2/r_1) \over 2{\pi}Lk}(단위 : [K/W]) $$
### $$ R^{'}_{t,cond} = {ln(r_2/r_1) \over 2{\pi}k}(단위 : [mK/W]) $$
### 복합 원통벽
> 접촉 열저항은 무시한다.

### **그림 3.8 복합 원통벽에 대한 온도분포**




### $$ q_r = {T_{\infty,1} - T_{\infty,A} \over R_{tot}} = UA(T_{\infty,1} - T_{\infty,A}) $$

```python
"""
example 3-6
critical insulation thickness
"""
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
from scipy.optimize import fmin, newton

h = 5
k = 0.055
ri = 10e-3/2
r = np.linspace(ri, 55e-3)
def R_fun(r):
    return np.log(r/ri)/(2*np.pi*k) + 1/(2*np.pi*r*h)
plt.plot(r-ri, R_fun(r))

# critical thickness
r_cr = fmin(R_fun, 0.01)[0]
print('critical thickness : ', r_cr - ri)

Rt_cr = R_fun(r_cr)
plt.plot(r_cr - ri, Rt_cr, 'o')
plt.text(r_cr - ri, Rt_cr + 0.1, 'critical thickness')

# practical thickness
r_pr = newton(lambda r: R_fun(r) - 1/(2*np.pi*ri*h), 0.03)
print('practical thickness : ', r_pr - ri)

Rt_pr = R_fun(r_pr)
plt.plot(r_pr - ri, Rt_pr, 'o')
plt.text(r_pr - ri, Rt_pr + 0.1, 'practical thickness')
plt.plot([0, r_pr - ri], [Rt_pr, Rt_pr], ':')

plt.xlabel('thickness, m')
plt.ylabel('thermal resistance, m-K/W')
plt.show()
```

## 3.4  1차원 전도 결과요약 




열에너지 발생이 없는 평면벽, 원통형벽, 구형벽에서의 1차원 정상상태전도에 대해 다루었다.
이들 세가지 형상에 대한 중요한 결과를 표로 정리했다.
각 경우마다 열방정식으로 시작하여 온도분포, 열유속, 열전달률 및 열저항에 대해 대응하는 식을 유도할 수 있다.


## 3.5열에너지 발생이 있는 전도
## 3.5.1 평면벽

>- 평면벽 1차원, 정상상태 전도
>
>- 단위체적당 에너지 발생이 균일
 (q'일정)
>
>- 열전도율 k 일정

### *비대칭 경계조건 

---

### 열방정식(Heat Equation) 

### $$ {d \over dx}(k{dT \over dx}) + q{'} = 0  \rightarrow {d^2T \over dx^2} +{ q{'} \over k} = 0 $$


### 일반해(General Solution)

### $$T(x) = -{q{'} \over 2k}x^2 + C_1x + C_2$$


>**경계조건(Boundary Condition)** : $T(-L) = T_{s,1} , T(L) = T_{s,2}$


### 온도분포

### $$ T(x) = {q{'}L^2 \over 2k}(1-{x^2 \over L^2}) + {T_s,_2 - T_s,_1 \over 2}{x \over L} + {T_s,_2 - T_s,_1 \over 2} $$



### *대칭 경계조건, 중앙 단면이 단열 

---

>양쪽 표면이 동일 온도 $T_s,1 = T_s,2 = T_s$ 로 유지되어 있을때

### 온도분포
### $$ T(x) = {q{'}L^2 \over 2k}(1-{x^2 \over L^2}) + {T_s} $$

> 위 식은 한쪽 면이 완전히 단열 되어 있고 다른 쪽 면은 고정된 온도$T_s$로 유지되어 있는 평면벽에도 적용할 수 있다.
 
### 표면 온도 ($T_s$)

>### Overall Energy balance on the wall
### $$ -E{'}_{out} + E{'}_s = 0$$

### $$ -hA_s(T_s - T_{\infty}) + q{'}A_sL = 0$$

### $$ T_s = T_{\infty} + {q{'}L \over h}$$

# 3.5.2 반지름 방향 계

### * Cylindrical

---

### 열방정식(Heat Equations)

### $${1 \over r}{d \over dr}(kr{dT \over dr}) + q{'} = 0 $$ 

### 일반해(General Solution)

### $$ T(r) = -({q \over 4k})r^2 + C_1ln_r + C_2$$


>**경계조건(Boundary Condition)** :${dT \over dr}|r = 0 ,T(r_0) = T_s$



### 온도분포
   
### $$T(r) = {q{'}r_0^2 \over 4k}(1-{r^2 \over r_0^2}) + T_s$$

### 표면온도 ($T_s$)
>### Overall energy balance

### $$ \dot{-E_{out}} + \dot{E_g} = 0 \rightarrow T_s = T_{\infty} + {\dot{q}r_0 \over 2h}$$

### $$ T_s = T_{\infty} + {q{'}r_0 \over 2h}$$

### * Spherical
---


### 열방정식(Heat Equations)

### $$ {1 \over r^2}{d \over dr}(kr^2{dT \over dr}) + q{'} = 0 $$

### 일반해(General Solution)

### $$ T(r) = -{\dot{q} \over 6k}r^2 - {C_1 \over r} + C_2$$



> **경계조건(Boundary Condition)** :$${dT \over dr}|r = 0, T(r_0) = T_s$$

### 온도분포

   
### $$T(r) = {\dot{q}r_0^2 \over 6k}(1-{r^2 \over r_0^2}) + T_s$$

### 표면온도 ($T_s$)

>### Overall energy balance

### $$ \dot{-E_{out}} + \dot{E_g} = 0 \rightarrow T_s = T_{\infty} + {\dot{q}r_0 \over 3h}$$
### $$ T_s = T_{\infty} + {\dot{q}r_0 \over 3h}$$



```python
"""
Table C.1 
One dimensional steady state solutions to the Heat Equation for 
plane, cylinder and spherical walls with uniform generation and 
asymmetrical surface conditions
"""

import numpy as np

pi = np.pi

def plane_temperature(x, L, T1, T2, qdot, k):
    d1 = qdot*L**2/(2*k)
    d2 = 1 - (x/L)**2
    d3 = (T2 - T1)/2
    d4 = (T2 + T1)/2
    T = d1*d2 + d3*x/L + d4
    return T

def plane_heatflux(x, L, T1, T2, qdot, k):
    qf = qdot*x - k/(2*L)*(T2 - T1)
    return qf

def cylinder_temperature(r, r1, r2, T1, T2, qdot, k):
    d1 = qdot*r2**2/(4*k)
    d2 = 1 - (r/r2)**2
    d3 = 1 - (r1/r2)**2
    d4 = np.log(r2/r)/np.log(r2/r1)
    T = T2 + d1*d2 - (d1*d3 + (T2-T1))*d4    
    return T

def cylinder_heatflux(r, r1, r2, T1, T2, qdot, k):
    d1 = qdot*r/2
    d2 = qdot*r2**2/(4*k)
    d3 = 1 - (r1/r2)**2
    d4 = r*np.log(r2/r1)
    qf = d1 - k*(d2*d3 + (T2-T1))/d4
    return qf

def cylinder_heatrate(r, r1, r2, T1, T2, qdot, k, L):
    d1 = qdot*r2**2/(4*k)
    d2 = 2*pi*L*k/np.log(r2/r1)
    d3 = 1 - (r1/r2)**2
    q = qdot*pi*L*r**2 - d2*(d1*d3 + (T2-T1))
    return q

def sphere_temperature(r, r1, r2, T1, T2, qdot, k):
    d1 = qdot*r2**2/(6*k)
    d2 = (1/r -1/r2)/(1/r1 - 1/r2)
    d3 = 1 - (r1/r2)**2
    T = T2 + d1*(1-(r/r2)**2) - (d1*d3 + (T2-T1))*d2    
    return T

def sphere_heatflux(r, r1, r2, T1, T2, qdot, k):
    d1 = qdot*r2**2/(6*k)
    d3 = 1 - (r1/r2)**2
    qf = qdot*r/3 - k*(d1*d3 + (T2-T1))/(r**2*(1/r1 - 1/r2))
    return qf
def sphere_heatrate(r, r1, r2, T1, T2, qdot, k):
    d1 = qdot*r2**2/(6*k)
    d2 = qdot**4*np.pi*r**3/3
    d3 = 1 - (r1/r2)**2    
    q = d2 - 4*np.pi*k*(d1*d3 + (T2-T1))/(1/r1 - 1/r2)
    return q


if __name__ == "__main__":
    import matplotlib.pyplot as plt
    plt.style.use('seaborn-whitegrid')
    from scipy.optimize import newton
    
    k = 5
    r1, r2 = 20e-3, 25e-3
    qdot = 5e6
    Tinf = 80
    h = 110
    Q = pi*(r2**2-r1**2)*qdot
    T1 = Tinf + Q/(2*pi*r1*h)

    def fun(T2):
        qf = cylinder_heatflux(r2, r1, r2, T1, T2, qdot, k)
        return qf
    T2 = newton(fun, 400)
    print(T1, T2)
    
    r = np.linspace(r1, r2)
    T = cylinder_temperature(r, r1, r2, T1, T2, qdot, k)
    q = cylinder_heatrate(r, r1, r2, T1, T2, qdot, k, 1)
    
    plt.plot(r, T)
    plt.show()
    
    plt.plot(r, q)
    plt.show()
```

## 3.6 확장표면에서 열전달


### 확장표면이란?

>확장표면(extended surface)이란 말은 일반적으로 고체 내에서 전도에 의한 열전달과 고체의 경계로부터 대류(또는 복사)에 의한 열전달을 포함하는 특별한 경우를 표현하는데 사용된다.
확장표면에서 경계로부터의 열전달 방향은 고체 내에서 열전달의 주방향에 수직이 된다.

>전도와 대류가 결합된 효과를 포함하는 다른 경우가 많이 있지만, 가장 흔한 응용은 특히 고체와 이에 인접한 유체 사이에서 열전달률을 증가시키기 위하여 확장표면을 사용하는 것이다. 이러한 확장표면을 휜(fin)이라 한다.

### 여러가지 휜 형상



## 3.6.1 일반적인 전도 해석

>- x방향 1차원 전도, 정상상태
>
>- 일정한 열전도율 
>
>- 표면으로부터 복사 무시 
>
>- 열발생이 없음 
>
>- 대류열전달계수 h가 표면 전체에서 일정하다고 가정한다. 




### 확장표면에 대한 에너지방정식의 일반적인 형태

### $${d^2T \over dx^2} + ({1 \over A_c}{dA_c \over dx}){dT \over dx} - {hp \over kA_c}(T - T_{\infty}) = 0 $$
> (p는 주위 길이)

# 3.6.2  횡단면적이 균일한 휜



### $${d^2T \over dx^2} + ({1 \over A_c}{dA_c \over dx}){dT \over dx} - {hp \over kA_c}(T - T_{\infty}) = 0 $$

>각각의 휜은 온도 T(0) = $T_b$ 인 바탕 표면에 붙어 있으며 온도 $T_{\infty}$인 유체 내로 확장되어 있다.   
>$ A_c $와 P는 일정하다. 따라서 ${dA_c \over dx} = 0 $이므로

### $$ {d^2T \over dx^2} + {hP \over kA_c}(T - T_{\infty}) = 0$$

### 초과 온도 (excess temperature)

### $$ {\theta}(x) = T(x) - T_{\infty}$$

>여기서 $T_{\infty}$는 일정하므로 ${d{\theta} \over dx} = {dT \over dx}$

### $$ {d^2{\theta} \over dx^2} + m^2{\theta} = 0 $$ 
### $$ [m^2 = {hP \over kA_c}]$$

### 일반해(General Solution)

### $${\theta}(x) = C_1e^{mx} + C_2e^{-mx}$$

>**경계조건(Boundary Condition)**   
>x = 0 에서 $${\theta}(0) = T_b - T_{\infty} = {\theta}_b $$   
>x = L 에서 $$ h{\theta}(L) = -k{d{\theta} \over dx}|_{x=L}$$




### 온도분포
### $$ {{\theta} \over {\theta}_b} = {cosh m(L-x) + (h/mk)sinh m(L-x) \over cosh mL + (h/mk)sinh mL}$$

 
    
    
### 휜 열전달률$(q_f)$
### $$ q_f = q_b = -kA_c{dT \over dx}|_{x=0} = -kA_c{d{\theta} \over dx}|_{x=0}$$
    

### $$ q_f = \sqrt{hPkA_c}{\theta}_b{sinh mL + (h/mk)cosh mL \over cosh mL + (h/mk)sinh mL}$$

```python
"""
Table 3.4 Temperature distribution and heat rates 
for fins of uniform cross section
"""

import numpy as np

def fun_m(h,P,k,A):
    return np.sqrt((h*P)/(k*A))

def fun_M(h,P,k,A,theta_b):
    return np.sqrt(h*P*k*A)*theta_b

def Case_A_temperature(x, m, L, k, h):    
    # convection
    d1 = np.cosh(m*(L-x)) + (h/(m*k))*np.sinh(m*(L-x))
    d2 = np.cosh(m*L) + (h/(m*k))*np.sinh(m*L)
    return d1/d2

def Case_A_heatrate(M, m, L, k, h):    
    # convection
    d1 = np.sinh(m*L) + (h/(m*k))*np.cosh(m*L)
    d2 = np.cosh(m*L) + (h/(m*k))*np.sinh(m*L)
    return M*d1/d2

def Case_B_temperature(x, m, L):    
    # adiabatic
    return np.cosh(m*(L-x))/np.cosh(m*L)

def Case_B_heatrate(M, m, L):    
    # adiabatic
    return M*np.tanh(m*L)

def Case_C_temperature(x, m, L, theta_L, theta_b):    
    # prescribed temperature
    d1 = (theta_L/theta_b)*np.sinh(m*x) + np.sinh(m*(L-x))    
    return d1/np.sinh(m*L)

def Case_C_heatrate(M, m, L, theta_L, theta_b):    
    # prescribed temperature
    q0 = M/np.sinh(m*L) * (np.cosh(m*L) - (theta_L/theta_b))
    qL = M/np.sinh(m*L) * (1 - np.cosh(m*L)*(theta_L/theta_b))
    return q0, qL

def Case_D_temperature(x, m):
    # infinite fin
    return np.exp(-m*x)

def Case_D_heatrate(M):
    # infinite fin
    return M
```

```python
"""
example 3-9
"""
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')

D = 5e-3
Tb = 100
h = 100
Tinf = 25

A = np.pi*D**2/4
P = np.pi*D

k_data = [('Cu', 398), ('Al', 180), ('SS', 14)]

x = np.linspace(0, 300e-3, 100)
T = np.empty_like(x)
for name, k in k_data:
    m = np.sqrt((h*P)/(k*A))
    T = Tinf + (Tb - Tinf)*np.exp(-m*x) # table 3-4, case D
    qf = np.sqrt(h*P*k*A)*(Tb - Tinf)
    L_inf = 2.65/m
    print(qf, L_inf)
    plt.plot(x, T, label=name)
    plt.plot([L_inf, L_inf], [20, 100], ':')

plt.xlabel('x, mm')
plt.ylabel('T, C')
plt.legend()
plt.show()    
```

# 3.6.3 휜의 성능

### 휜유용도(Fin Effectiveness)

>휜으로의 열전달율과 휜이 없을 때의 열전달율의 비로서 정의된다.

### $$ {\epsilon}_f = {q_f \over hA_{c,b}{\theta}_b} $$

### 휜저항(Fin Resistance)

### $$ R_{t,f} = {{\theta}_b \over q_f}$$

### 면유체에 노출된 바탕면으로 인한 열저항

### $$ R_{t,b} = {1 \over hA{c,b}}$$

위 두식을 결합하면

### $$ {\epsilon}_f = {R_{t,b} \over R_{t,f}} $$
> 휜의 열전달을 증가 시키기 위한 것이라면 휘의 열저항이 노출된 바탕의 열저항보다 커서는 안된다.

### 휜 효율(Fin Efficiency)

>실제 열전달율를 가능한 최대 열전달율로 나눈 것

### $$ n_f = {q_f \over q_{max}} = {q_f \over hA_f(T_b-T_{\infty})} = {q_f \over hA_f{\theta}_b}  $$

>(여기서 A_f는 휜의 표면적)

## 3.6.5 종합 표면효율

### 종합 표면 효율(overall surface efficiency)${\eta}_0$
>휜의 배열과 휜이 부착되어 있는 바탕 표면의 특성을 나타낸다.




### $$ {\eta}_0 = {q_t \over q_{max}} = {q_t \over hA_t{\theta}_b}$$


### $ \Rightarrow A_t = NA_f + A_b $
>$A_t$ : 총 표면적  
>$A_f$ : 휜의 표면적   
>$A_b$ : 기초 표면적

### 총 열전달률

### $$ q_t = N{\eta}_fhA_f{\theta}_b + hA_b{\theta}_b \equiv \eta_0hA_t\theta_b $$
   

>대류열전달계수 h를 휜 표면과 기초 표면에서 같다고 가정하면

   
### $$ q_t = h[N{\eta}_fhA_f{\theta}_b + (A_t - NA_f)]{\theta}_b = hA_t[1- {NA_f \over A_t}(1-{\eta}_f)]{\theta}_b$$
  

### $$ {\eta}_0 = 1- {NA_f \over A_t}(1-{\eta}_f)$$


### 휜 배열의 열저항 
>휜에서 전도와 대류에 의한 그리고 기초 표면으로부터 대류에 의한 평행 열흐름 통로에 대한 유효저항이다.



### $$ R_{t,0} = {{\theta}_b \over q_t} = {1 \over {\eta}_0hA_t}$$

### 열접촉저항 $R_{t,e}$가 존재할 때



### $$ q_t =  {\eta}_{0(c)}hA_t{\theta}_b = {{\theta}_b \over R_{t,0(c)} }$$

### $$ R_{t,0(c)} = {{\theta}_b \over q_t} = {1 \over{\eta}_{0(c)}hA_t } $$

### $${\eta}_{0(c)} = 1 - {NA_f \over A_t}(1-{{\eta}_f \over C_1}) $$

### $$ C_1 = 1+ {\eta}_{f}hA_f(R{''}_{t,c}/A_{c,b}) $$

```python
import numpy as np
from scipy.special import i0, i1, k0, k1
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')


pi = np.pi


def overall_effeciency(eta_f, N, Af, At, Rtc=0, h=0, Ab=0):
    if Rtc == 0:
        C1 = 1
    else:
        C1 = 1 + eta_f*h*Af*Rtc/Ab
    return 1 - N*Af/At*(1 - eta_f/C1)


def straight_rectangular(k, h, L, t, w=1, is_convection_tip=True):
    if is_convection_tip: 
        Lc = L + t/2
    else:
        Lc = L
    m = np.sqrt((2*h)/(k*t))
    Af = 2*w*Lc
    Ap = t*L
    eta = np.tanh(m*Lc)/(m*Lc)
    return eta, m, Af, Ap


def straight_triangular(k, h, L, t, w=1):
    m = np.sqrt((2*h)/(k*t))
    Af = 2*w*np.sqrt(L**2 + (t/2)**2)
    Ap = t*L/2
    eta = 1/(m*L)*i1(2*m*L)/i0(2*m*L)
    return eta, m, Af, Ap


def straight_parabolic(k, h, L, t, w=1):
    m = np.sqrt((2*h)/(k*t))
    C1 = np.sqrt(1 + (t/L)**2)
    Af = w*(C1*L + L**2/t*np.log(t/L + C1))
    Ap = t/3*L
    eta = 2/(np.sqrt(4*(m*L)**2 + 1) + 1)
    return eta, m, Af, Ap


def annular_fin(k, h, r1, r2, t, is_convection_tip=True):
    if is_convection_tip:
        r2c = r2 + t/2
    else:
        r2c = r2
    m = np.sqrt((2*h)/(k*t))
    Af = 2*pi*(r2c**2 - r1**2)
    V = pi*(r2**2 - r1**2)*t
    C2 = (2*r1/m)/(r2c**2 - r1**2)
    d1 = k1(m*r1)*i1(m*r2c) - i1(m*r1)*k1(m*r2c)
    d2 = i0(m*r1)*k1(m*r2c) + k0(m*r1)*i1(m*r2c)
    eta = C2*d1/d2
    return eta, m, Af, V


def pin_fin_rectangular(k, h, L, D, is_convection_tip=True):
    if is_convection_tip:
        Lc = L + D/4
    else:
        Lc = L
    m = np.sqrt((4*h)/(k*D))
    Af = pi*D*Lc
    V = pi*D**2/4 * L
    eta = np.tanh(m*Lc)/(m*Lc)
    return eta, m, Af, V


if __name__ == "__main__":

    k = 100
    h = 10
    t = 1e-3

    L = np.linspace(1e-3, 250e-3)
    x = np.empty_like(L)
    eta = np.empty_like(L)

    for i in range(len(L)):
        eta[i], m, Af, Ap = straight_rectangular(k, h, L[i], t)
        Lc = L[i] + t/2
        x[i] = Lc**(3/2)*np.sqrt(h/(k*Ap))
    plt.plot(x, eta, label='rectangular')

    for i in range(len(L)):
        eta[i], m, Af, Ap = straight_triangular(k, h, L[i], t)
        x[i] = L[i]**(3/2)*np.sqrt(h/(k*Ap))
    plt.plot(x, eta, label='triangular')

    for i in range(len(L)):
        eta[i], m, Af, Ap = straight_parabolic(k, h, L[i], t)
        x[i] = L[i]**(3/2)*np.sqrt(h/(k*Ap))
    plt.plot(x, eta, label='parabolic')

    plt.xlim([0, 2.5])
    plt.legend()
    plt.xlabel(r'$L_c^{3/2} (h/k A_p)^{1/2}$')
    plt.ylabel(r'$\eta$')
    plt.show()
```
